## Cumulative Density Function
The cumulative distribution function (CDF) FX(x) describes the probability that a random variable X with a given probability distribution will be found at a value less than or equal to x.

## Kernel Density estimation
Kernel density estimation is a really useful statistical tool with an intimidating name. Often shortened to KDE, its a technique that lets you create a smooth curve given a set of data.
This can be useful if you want to visualize just the shape of some data, as a kind of continuous replacement for the discrete histogram. It can also be used to generate points that look like they came from a certain dataset - this behavior can power simple simulations, where simulated objects are modeled off of real data.

```python
import pandas as pd
df= pd.read_csv('titanic_data.csv')
df.head()
```

```python
df['Fare'].plot.kde()
```

```python
import seaborn as sns
sns.kdeplot(data=df['Age'],shade=True)
```
