# Seaborn Learning
## Distribution plot
- distplot
- joinplot
- pairplot

```python
import seaborn  as sns
df= sns.load_dataset('tips')
df.head()
```

### correlation with heatmap
A correlation with heatmap is coloured matrix to show the 2D correlation matrix with two discrete dimension.It is very important for feature selection

```python
df.dtypes
```

```python
df.corr()
```

```python
sns.heatmap(df.corr()) # check what is correlation and what it signify? 
```

### jointplot
A jointplot allow the study of relationship of two numeric feature
### Univariante analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### pairplot
it will do permutation and combination of all the features

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
## it is histogram
sns.distplot(df.tip)
```

```python
sns.distplot(df.tip,kde=False) #kde false will provide the count
```

## Categorical plot
- boxplot
- voilinplot
- countplot
- barplot

###   Countplot

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

### Barplot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

### Boxplot
A box plot and whisker plot shows the information of five number summary

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=df)
```

### Violinplot
violin plot is the graph which give the information of both box plot and kernel density estimation(kde).

```python
sns.violinplot(x='day',y='total_bill',data=df)
```
